```python
from bs4 import BeautifulSoup
import requests
response = requests.get('https://www.freelancer.com/job/')
soup = BeautifulSoup(response.text, 'html.parser')
all_numbers = 0
section = soup(id="category-6")
types_dictionary = {}
for lis in soup("li"): #loop for collecting data
    li_number = 0
    li_text=""
    first = True
    for ais in lis("a"): #collects all numbers from subcategories and assignes them to primary categories
        if(ais.text.rfind("(") != -1):
            potential_number = ais.text[ais.text.rfind("(")+1:ais.text.rfind(")")]
            if(potential_number.isdigit()):
                number = int(potential_number)
                li_number+=number
                all_numbers+=number
                if(first):
                    first = False
                    li_text = ais.text
    types_dictionary[li_text]=li_number #adds category with collected numbers to dictionary
#transform dictionary into sorted list
sorted_list = sorted(types_dictionary.items(), key=lambda item: item[1], reverse=True)
print(all_numbers)   
for r in sorted_list: #loop for printing data
    print(r[0] + str(r[1]))
```
